It's particularly effective for classification problems. </div>

<div style="text-align: left; background-color:; font-family: Trebuchet MS; color: #00008B; padding: 12px; line-height:1;border-radius:1px; margin-bottom: 0em; text-align: MIDDLE; font-size: 25px;border-style: solid;border-color: dark green"> Imagine you have two groups of data points (e.g., cats and dogs) that you want to separate with a straight line. </div>

## 1) SVM helps you find the best line (called a "hyperplane") that separates these groups with the largest possible gap between them.
## 2)This line should be positioned in such a way that it maximizes the distance from the nearest data points of each group to the line.
## 3)These nearest data points are called "support vectors," hence the name "Support Vector Machine."

## Here are some key points:

1) Linear Separation: SVM works well when your data can be separated by a straight line (for 2D data), a plane (for 3D data), or a hyperplane (for higher-dimensional data).

2) Kernel Trick: SVM can handle non-linear data by using a mathematical function called a "kernel" that transforms the data into a higher-dimensional space where a hyperplane can separate it. Common kernels include the Radial Basis Function (RBF) and polynomial kernels.

3) Margin: The distance between the support vectors and the hyperplane is called the "margin." SVM aims to maximize this margin, which helps the model generalize well to new, unseen data.

4) C Parameter: SVM has a parameter called "C" that controls the trade-off between maximizing the margin and minimizing the classification error. A smaller C value will prioritize a larger margin, potentially allowing some misclassifications, while a larger C value will prioritize correct classifications, potentially leading to a smaller margin.

5) Multi-Class Classification: SVM can be extended to handle multi-class classification problems by using techniques

# Train Using Support Vector Machine (SVM)

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target','flower_name'], axis='columns')
y = df.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict([[4.8,3.0,1.5,0.3]])
```
